```python
import pandas as pd
```

```python
rating = pd.read_csv('Amazon - Movies and TV Ratings.csv')
```

```python
rating.head()
```

```python
rating.set_index('user_id', inplace=True)
```

```python
#Handling of missing data 
missing_column = rating.isnull()

for missing in missing_column.columns.values.tolist():
    print(missing)
    print(missing_column[missing].value_counts())
```

### Which movies have maximum views/ratings?

```python
max_ratting = {}
for movies in rating.columns.values.tolist():
    max_ratting[movies] = rating[rating[movies] == rating[movies].max()][movies].count()
    
print(max(max_ratting.items(), key=lambda k: k[1]))
```

### What is the average rating for each movie? Define the top 5 movies with the maximum ratings.


```python
columns = ['Mean', 'Review Count']
data = []
index = []
for movies in rating.columns.values.tolist():
    data.append([rating[movies].mean(), rating[rating[movies] >= rating[movies].mean()][movies].count()])
    index.append(movies)

avg_rating = pd.DataFrame(data, index=index, columns=columns)
```

```python
avg_rating.head()
```

```python
data = avg_rating[avg_rating['Mean'] >= 4]
print(data.sort_values(by=['Review Count'], ascending=False).head())
```

### Define the top 5 movies with the least audience.
