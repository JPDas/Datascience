### Linear Regression

```python
X = house[['bedrooms', 'bathrooms', 'sqft_living','sqft_lot', 'floors','sqft_above', 'sqft_basement', 'yr_built', 
           'condition', 'zipcode']]
```

```python
y = house['price']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Traning and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=7)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
#Fit the model
model1.fit(X_train, y_train)
```

```python
#Prediction
y_pred1 = model1.predict(X_test)
```

```python
#Root mean square error(RMSE)
from sklearn.metrics import mean_squared_error
```

```python
mean_sq_err = mean_squared_error(y_test, y_pred1)
```

```python
mean_sq_err
```

```python
RMSE = np.sqrt(mean_sq_err)
```

```python
RMSE
```
