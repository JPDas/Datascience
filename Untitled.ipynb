```python
train_loader = torch.utils.data.DataLoader(dataset = train_embeddings,
                                     batch_size = 32,
                                     shuffle = True)
```

```python
import torch
import torch.nn as nn
```

```python
# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
class AutoEncoder(nn.Module):
    def __init__(self, embedding_len, encoder_dim):
        super().__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(embedding_len, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, encoder_dim),
            
        )
        
        self.decoder = torch.nn.Sequential(
            
            torch.nn.Linear(encoder_dim, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, embedding_len),
            torch.nn.Sigmoid()
        )
        
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
```

```python
# create a model from `AE` autoencoder class
# load it to the specified device, either gpu or cpu
model = AutoEncoder(768, 32).to(device)

# Validation using MSE Loss function
loss_function = torch.nn.MSELoss()
 
# Using an Adam Optimizer with lr = 0.1
optimizer = torch.optim.Adam(model.parameters(),
                             lr = 1e-1,
                             weight_decay = 1e-8)
```

```python
epochs = 20
outputs = []
losses = []
for epoch in range(epochs):
    for batch in train_loader:
        
        for text in batch:
       
            # Output of Autoencoder
            reconstructed = model(text)

            # Calculating the loss function
            loss = loss_function(reconstructed, text)

            # The gradients are set to zero,
            # the gradient is computed and stored.
            # .step() performs parameter update
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Storing the losses in a list for plotting
            losses.append(loss)
    outputs.append((epochs, text, reconstructed))
```

```python
np_losses = []
for loss in losses:
    np_losses.append(loss.detach().cpu().numpy())
```

```python
import matplotlib.pyplot as plt

# Defining the Plot Style
plt.style.use('fivethirtyeight')
plt.xlabel('Iterations')
plt.ylabel('Loss')
 
# Plotting the last 100 values
plt.plot(np_losses[-100:])
```

```python
class VariationalEncoder(nn.Module):
    def __init__(self, embedding_len, latent_dims):
        super(VariationalEncoder, self).__init__()
        self.fc = torch.nn.Sequential(
            torch.nn.Linear(embedding_len, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 64),
            torch.nn.ReLU())
        self.linear3 = nn.Linear(64, latent_dims)
        self.linear4 = nn.Linear(64, latent_dims)

        self.N = torch.distributions.Normal(0, 1)
        self.N.loc = self.N.loc.cpu() # hack to get sampling on the GPU
        self.N.scale = self.N.scale.cpu()
        self.kl = 0

    def forward(self, x):
        
        x = self.fc(x)
        
        mu =  self.linear3(x)
        sigma = torch.exp(self.linear4(x))
        z = mu + sigma*self.N.sample(mu.shape)
        self.kl = (sigma**2 + mu**2 - torch.log(sigma) - 1/2).sum()
        return z
```

```python
class VariationalAutoencoder(nn.Module):
    def __init__(self, embedding_len, latent_dims):
        super(VariationalAutoencoder, self).__init__()
        self.encoder = VariationalEncoder(embedding_len, latent_dims)
        self.decoder = torch.nn.Sequential(
            
            torch.nn.Linear(latent_dims, 64),
            torch.nn.ReLU(),
            torch.nn.Linear(64, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, embedding_len),
            torch.nn.Sigmoid()
        )

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)
```

```python
# create a model from `AE` autoencoder class
# load it to the specified device, either gpu or cpu
model = VariationalAutoencoder(768, 32).to(device)

# Validation using MSE Loss function
loss_function = torch.nn.MSELoss()
 
# Using an Adam Optimizer with lr = 0.1
optimizer = torch.optim.Adam(model.parameters(),
                             lr = 1e-1,
                             weight_decay = 1e-8)

```

```python
epochs = 20
outputs = []
losses = []
val_loss = 0
n_epochs_stop = 6
epochs_no_improve = 0
early_stop = False
min_val_loss = np.Inf

for epoch in range(epochs):
    for batch in train_loader:
        
        for text in batch:
       
            # Output of Autoencoder
            reconstructed = model(text)

            # Calculating the loss function
            loss = loss_function(reconstructed, text)

            # The gradients are set to zero,
            # the gradient is computed and stored.
            # .step() performs parameter update
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Storing the losses in a list for plotting
            losses.append(loss)
            
            val_loss += loss
            val_loss = val_loss / len(train_loader)
            # If the validation loss is at a minimum
            if val_loss < min_val_loss:
                # Save the model
                #torch.save(model)
                epochs_no_improve = 0
                min_val_loss = val_loss

            else:
                epochs_no_improve += 1
                
            if epoch > 5 and epochs_no_improve == n_epochs_stop:
                print('Early stopping!' )
                early_stop = True
                break
            else:
                continue
        if early_stop:
            print("Stopped")
            break
            
    if early_stop:
        print("Stopped")
        break
        
    outputs.append((epochs, text, reconstructed))
```

## LSTM based Autoencoder

```python
def create_dataset(sequences):
    dataset = [torch.tensor(s).unsqueeze(1) for s in sequences] 
 
    n_seq, seq_len, n_features = torch.stack(dataset).shape # n_seq,4,1
    return n_seq, seq_len, n_features
```

```python
_, seq_len, n_features = create_dataset(train_embeddings)
```

```python
class Encoder(nn.Module):
    def __init__(self, batch_size, seq_len, n_features, embedding_dim, n_layers):
        super(Encoder, self).__init__()

        self.seq_len = seq_len
        self.batch_size = batch_size
        self.n_features = n_features
        self.hidden_dim = 2 * embedding_dim

        self.rnn1 = nn.LSTM(
          input_size=n_features,
          hidden_size=self.hidden_dim,
          num_layers=n_layers,
          batch_first=True  # True = (batch_size, seq_len, n_features)
                            # False = (seq_len, batch_size, n_features) 
                            #default = false
        )
        self.rnn2 = nn.LSTM(
          input_size=self.hidden_dim,
          hidden_size=embedding_dim,
          num_layers=n_layers,
          batch_first=True
        )

    def forward(self, x):
        #(4,1)
        x = x.reshape((self.batch_size, self.seq_len, self.n_features)) 
        # (batch, seq, feature)   #(1,4,1) 
        x, (_, _) = self.rnn1(x) #(1,4,256) 
        x, (hidden_n, _) = self.rnn2(x)
        # x shape (1,4,128)
        # hidden_n (1,1,128)
        return hidden_n.reshape((self.n_features, self.embedding_dim)) #(1,128)
```

```python
class Decoder(nn.Module):
    def __init__(self, batch_size, seq_len, n_features, embedding_dim, n_layers):
        super(Decoder, self).__init__()

        self.seq_len = seq_len
        self.hidden_dim = 2 * embedding_dim
        self.n_features = n_features
        self.n_layers = n_layers
        self.batch_size = batch_size
        self.embedding_dim = embedding_dim

        self.rnn1 = nn.LSTM(
        input_size=self.embedding_dim,
        hidden_size=self.embedding_dim,
        num_layers=self.n_layers,
        batch_first=True
        )
        self.rnn2 = nn.LSTM(
        input_size=self.embedding_dim,
        hidden_size=self.hidden_dim,
        num_layers=self.n_layers,
        batch_first=True
        )
        self.output_layer = nn.Linear(self.hidden_dim, n_features)

    def forward(self, x):
        # x shape(1,128)
        x = x.repeat(self.seq_len, self.batch_size) # (4, 128)

        #x = x.reshape((n_features, self.seq_len, self.embedding_dim)) # 1,4,128
        x, (hidden_n, cell_n) = self.rnn1(x) #(1,4,128)
        x, (hidden_n, cell_n) = self.rnn2(x) #(1,4,256)
        x = x.reshape((self.seq_len, self.hidden_dim)) #(4,256)

        return self.output_layer(x) # (4,1)
```

```python
class RecurrentAutoencoder(nn.Module):
    def __init__(self, batch_size, seq_len, n_features, embedding_dim, n_layers):
        super(RecurrentAutoencoder, self).__init__()
        self.seq_len = seq_len
        self.n_features = n_features

        self.encoder = Encoder(batch_size, seq_len, n_features, embedding_dim, n_layers).to(device)
        self.decoder = Decoder(batch_size, seq_len, n_features, embedding_dim, n_layers).to(device)
    def forward(self, x):
        x = self.encoder(x)

        x = self.decoder(x)
        return x
```

```python
model = RecurrentAutoencoder(32, seq_len, n_features, 32, 1)
```
