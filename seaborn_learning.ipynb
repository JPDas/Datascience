## Seaborn
Seaborn is a most popular visualization tool which provides the attractive data visualization for pattern discovery. It internally uses matplotlib library.

We will discuss the univariate or bivariate distributions and for comparing them between subsets of data.

To install the seaborn, we can use pip or conda

pip install seaborn

### Import dependencies

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### Seaborn has inbuilt dataset. We will import the tips dataset and then we will analyse them using univariate and bivariate.

```python
#Inbuilt seaborn dataset
tips = sns.load_dataset('tips')
tips.head()
```

### Numerical Plot

Plot using numerical variables/features.

#### 1. Distribution plot
Univariate distribution, displaying the distribution of the data and range of the data.

The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions.

```python
sns.distplot(tips['total_bill'],bins=20, color='red')
```

##### 2. Joint Plot
Seaborn’s jointplot displays a relationship between 2 variables (bivariate).It is similar to scatter plot.

```python
sns.jointplot(x='total_bill',y='tip', data=tips)
```

#### 3. Pair plot
Joint plot is among 2 variable plot, but pair plot will draw entire variable in the dataset.

```python
sns.pairplot(tips,hue='sex')
```

#### 4. Rug plot

Plots all datapoints in an array as sticks on an axis

```python
sns.rugplot(tips['total_bill'])
```

##### 5. KDE plot

```python
#KDE plot
sns.kdeplot(tips['total_bill'])
```

### Categorical plot

#### 1. Bar plot

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std)
```

#### 2. Count plot

```python
sns.countplot(x='sex', data=tips)
```

#### 3. Box plot

Box plot provides percentile of data.

```python
#Box plot shows the percentile of data
sns.boxplot(x='day', y='total_bill', data=tips)
```

#### 4. Violin plot

Combination of histogram, PDF and box plot.

```python
#Violin plot
sns.violinplot(x='day', y = 'total_bill', data=tips)
```

#### 5. Strip plot
Draw a scatter plot where one variable is categorical.

```python
#strip plot
sns.stripplot(x = 'day', y='total_bill', data=tips, jitter=True)
```

#### 6. Swarm plot

Draws the scatter plot with non overlapping points.

```python
#swarmplot
#draw a scatter plot with non-overlapping points
sns.swarmplot(x='day', y='total_bill', data=tips)
```

#### 7. Factor Plot

```python
#Factor plot
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

### Matrix plot

```python
#Heatmap is an example of matrix plot
#To convert the dataset in to matrix, we need to create in below 2 ways
#1. Pivot table
#2. Correlation

tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
#Grids
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris).map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
